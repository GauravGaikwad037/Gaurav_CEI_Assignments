## Basic Data Exploration and Cleaning using Pandas

**Objective:**
Perform data loading, exploration, cleaning, preprocessing, and feature engineering on the provided shopping dataset using Python and Pandas.

# Firstly I Imported the Required Libraries

```python

import pandas as pd
import os
```

# Then I Load All CSV Files

```python
folder_path = "data/raw"
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]
print("Total CSV files found:", len(csv_files))
print(csv_files)
```

# Then I Combine All CSV Files into One DataFrame

```python

dataframes = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)

print("All CSV files combined successfully!")
```

# Checking Missing Values

```python
print("Missing Values in Each Column:")
combined_df.isnull().sum()
```

# Then I Fill The Missing Values

```python
for column in combined_df.columns:
    if combined_df[column].dtype == "object":
        combined_df[column] = combined_df[column].fillna("Unknown")
    else:
        combined_df[column] = combined_df[column].fillna(0)

print("Missing values handled successfully!")
```

# Removing Duplicate Records

```python
before = combined_df.shape[0]

combined_df = combined_df.drop_duplicates()

after = combined_df.shape[0]

print("Duplicates removed:", before - after)
```

# Selecting Required Columns

```python
selected_columns = combined_df[
    ["product_id", "title", "final_price", "rating", "category"]
]

selected_columns.head()
```

# Filtering Products with Rating 4 or Above

```python
filtered_products = combined_df[combined_df["rating"] >= 4]

filtered_products.head()
```

# Creating New Columns

```python
combined_df["quantity"] = 1

combined_df["total_amount"] = (
    combined_df["final_price"] * combined_df["quantity"]
)

combined_df[
    ["final_price", "quantity", "total_amount"]
].head()
```

# Saving the Cleaned Dataset

```python
output_path = "data/cleaned/cleaned_shopping_dataset.csv"

combined_df.to_csv(output_path, index=False)

print("Cleaned dataset saved successfully!")
```
